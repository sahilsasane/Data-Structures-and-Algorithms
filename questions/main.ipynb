### combine

```python
import pandas as pd
import os
import glob
```

```python
csvs = glob.glob('./apple/*.csv', recursive=True)
csvs
```

```python

df1 = pd.read_csv(csvs[0], encoding='utf-8')
df2 = pd.read_csv(csvs[1], encoding='utf-8')
df3 = pd.read_csv(csvs[2], encoding='utf-8')
df4 = pd.read_csv(csvs[3], encoding='utf-8')
```

```python
combined = pd.concat([df1, df2, df3, df4], axis=0, ignore_index=True)
```

```python
combined
```

```python
combined = combined.drop_duplicates(subset=['ID'], keep='first')
```

```python
# sort by frequency descending
combined = combined.sort_values(by=['Frequency'], ascending=False)
```

```python
combined
```

```python
combined.to_csv("./apple/combined.csv", index=False, encoding='utf-8')
```

### combined analysis

```python
google = pd.read_csv('./google/combined.csv', encoding='utf-8')
facebook = pd.read_csv('./facebook/combined.csv', encoding='utf-8')
amazon = pd.read_csv('./amazon/combined.csv', encoding='utf-8')
```

```python
combined = pd.concat([combined, google, facebook, amazon], axis=0, ignore_index=True)
combined
```

```python
combined.drop_duplicates(subset=['ID'], keep='first', inplace=True)
combined
```

```python
combined.to_csv("./combined.csv", index=False, encoding='utf-8')
```
